```python
import librosa
import matplotlib.pyplot as plt

# Load original audio
y_orig, sr_orig = librosa.load('../Covid_19 Project/Audio Prepro/00a2faca-e1f2-4848-9afe-058f949d3252.wav')

# Load preprocessed audio
y_preprocessed, sr_preprocessed = librosa.load('../Covid_19 Project/Peak audio/00a2faca-e1f2-4848-9afe-058f949d3252.wav')

# Plot original audio signal
fig, ax = plt.subplots(nrows=2, sharex=True, sharey=True, figsize=(10, 10))
ax[0].set(title='Original Audio Signal')
ax[0].plot(y_orig)

# Plot preprocessed audio signal
ax[1].set(title='Preprocessed Audio Signal')
ax[1].plot(y_preprocessed)

plt.show()
```
